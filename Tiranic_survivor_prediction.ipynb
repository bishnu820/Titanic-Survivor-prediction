```python
data=raw_data.copy()
```

```python
data=data.drop(["Ticket","Cabin","PassengerId","Name"],axis=1)
```

```python
data["Family"]=data["SibSp"]+data["Parch"]
```

```python
data1=data.drop(["SibSp","Parch"],axis=1)
```

```python
data1.describe(include="all")
```

```python
from sklearn.preprocessing import LabelEncoder
Encoder=LabelEncoder()
data1["Sex"]=Encoder.fit_transform(data1["Sex"])
data1["Embarked"]=Encoder.fit_transform(data1["Embarked"])
```

```python
data1["Age"]=data1["Age"].fillna(data1.Age.mean())
```

```python
data1
```

```python
data1=data1.dropna(axis=0)
```

```python
data1.describe()
```

```python
inputs=data1.drop(["Survived"],axis=1)
targets=data1.Survived
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
model.fit(inputs,targets)
```

```python
model.score(inputs,targets)
```
